###  Grain size analysis in python using the watershed
https://imagej.net/plugins/classic-watershed
Classic watershed is an method to perform watershed segmentation of grayscale 2D/3D images images using flooding simulations.
the basic idea consists of considering the input image as topographic surface and placing a water source in each regional minimum of its relief. next the entire relief is flooded from the sources and dams are placed where the different water sources meet.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from skimage import io, color, measure
from skimage.segmentation import clear_border
```

read the image

```python
img_path = './images/grains/grains2.jpg'
img1 = cv2.imread(img_path, -1)
img = cv2.imread(img_path, 0)
```

```python
pixels_to_um = 0.5
```

using Otsu threshold, get a binary image

```python
ret, thresh = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
ret
```

define a kernel, which is used for morphological openning to smooth the image boundary

```python
kernel = np.ones((3,3), np.uint8)
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=1)
io.imshow(opening)
```

for the grain size calculation, the grain in the picture cut by the border should not be considered, to exclude the grains on the border, use a skimage.segmentation.clear_border function

```python
opening = clear_border(opening)
```

identify the pixels definitely belong to foreground and background

```python
sure_bg = cv2.dilate(opening, kernel, iterations =2)
io.imshow(sure_bg)
```

Using the distance transformation method to find the absolute foreground

The distance transform operator generally takes binary images as input, the gray level intensities of the points inside the foreground regions are changed to distance their respective distances from the closest 0 value(boundary).

My interpretation of use the distance transform is to get a derived representation of a binary image, and convinient to apply laplacian filtering for image sharpening.

In this case, we are trying to apply the watershed method, so here after the distance transform, we build up the mountains and valleys

```python
dist_transform = cv2.distanceTransform(opening, cv2.DIST_L2, 3)
io.imshow(dist_transform)
```

from the distance transform image, apply threshold again, to get the absolute foreground

```python
ret_fg, sure_fg = cv2.threshold(dist_transform, 0.2*dist_transform.max(), 255, 0)
io.imshow(sure_fg)
```

check the np array data type

```python
sure_fg.dtype
```

```python
sure_fg = sure_fg.astype(np.uint8)
sure_fg.dtype
```

after getting the absolute background and foreground, there are pixels that are not known yet, show the unknown 

```python
unknown = cv2.subtract(sure_bg, sure_fg)
io.imshow(unknown)
```

watershed function need markers, cv2.connectedComponents is used here

```python
_, markers = cv2.connectedComponents(sure_fg)
markers
io.imshow(markers==2)
```

from the markers image shown above, the background is 0, and the foreground is labeled

we need to distinguish the makers background, so we add 10 to markers image, shift up everthing by 10

```python
markers += 10
io.imshow(markers)
```

set the unknown part as zero

```python
markers[unknown==255]=0
```

proceed the watershed

```python
markers = cv2.watershed(img1, markers)
img1[markers == -1] = [0, 255, 255]
img2 = color.label2rgb(markers, bg_label=0)
io.imshow(img1)
```
