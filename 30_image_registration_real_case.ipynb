```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

read the image

```python
img_path_1 = './images/monkey_distorted.jpg'
img_path_2 = './images/monkey.jpg'
img1_color = cv2.imread(img_path_1, -1)
img2_color = cv2.imread(img_path_2, -1)
```

change the color image to gray

```python
img1 = cv2.cvtColor(img1_color, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2_color, cv2.COLOR_BGR2GRAY)
```

initialize ORB

```python
orb = cv2.ORB_create(50)
kp1, des1 = orb.detectAndCompute(img1, None)
kp2, des2 = orb.detectAndCompute(img2, None)
```

match the keypoints, with bruteforce method

```python
matcher = cv2.DescriptorMatcher_create(cv2.DESCRIPTOR_MATCHER_BRUTEFORCE_HAMMING)
```

match the descripter with the matcher created

```python
matches = matcher.match(des1, des2, None)
```

sort the matches, based on the distance. distance could be interpreted as the residue of matching

```python
matches = sorted(matches, key=lambda x:x.distance )
```

initialize the empty array for the keypoints 

```python
points1 = np.zeros((len(matches),2),dtype=np.float32)
points2 = np.zeros((len(matches),2),dtype=np.float32)
```

replace the zero list with the keypoint index

```python
for i, match in enumerate(matches):
    points1[i,:] = kp1[match.queryIdx].pt
    points2[i,:] = kp2[match.trainIdx].pt
```

inspect the homography matrix, from the point1, point2 list. select only the inliners

```python
h, mask = cv2.findHomography(points1, points2, cv2.RANSAC)
```

use the homography, first get the dimensions of the reference image (not distorted), use the warpPerspective, transform the distorted image to the un-distorted image dimension, or I interprete it as tilt my perspective to match with the transformation.  

```python
height, width, channels = img2_color.shape
img1_Reg = cv2.warpPerspective(img1_color,h,(width, height))
```
