## Block matching and 3D filtering (BM3D) for denoising



```python
# pip install bm3d
```

```python
import bm3d
from skimage import io, img_as_float
from matplotlib import pyplot as plt
import cv2
```

```python
noisy_img = img_as_float(io.imread("images/BSE_Google_noisy.jpg"))
bm3d_denoised = bm3d.bm3d(noisy_img, sigma_psd=0.2, stage_arg=bm3d.BM3DStages.ALL_STAGES)
```

```python
io.imshow(bm3d_denoised)
```
