# digital image filtering 

```python
import numpy as np
from skimage import io, img_as_float
import matplotlib.pyplot as plt
import cv2
```

```python
img = io.imread('./images/BSE_Google_noisy.jpg',as_gray=True)
```

```python
img = img_as_float(img)
```

```python
io.imshow(img)
```

```python
kernel = np.ones((5,5), np.float32)/25
kernel
```

```python
gaussian_kernel = np.array([[1/16,1/8,1/16],
                           [1/8,1/4,1/8],
                           [1/16,1/8,1/16]])
gaussian_kernel
```

```python
gaussian_kernel.sum()
```

```python
conv_using_cv2 = cv2.filter2D(img, -1, kernel, borderType=cv2.BORDER_CONSTANT)
```

```python
io.imshow(conv_using_cv2)
```

image is denoised and blurred

```python
conv_using_cv2 = cv2.filter2D(img, -1, gaussian_kernel, borderType=cv2.BORDER_CONSTANT)
```

```python
io.imshow(conv_using_cv2)
```

scipy has same function for convolution, just for information

```python
from scipy.signal import convolve2d
from scipy.ndimage.filters import convolve
```

```python
conv_using_scipy_signal = convolve2d(img, kernel, mode='same')
conv_using_scipy_ndimage = convolve(img, kernel, mode='constant', cval=0.0)
```

```python
io.imshow(conv_using_scipy_signal)
```

```python
io.imshow(conv_using_scipy_ndimage)
```
