### Ridge filter to detect tube 

```python
from skimage import io, filters, feature
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
import cv2
```

The function that we need for this lesson

```python
from skimage.filters import meijering, sato, frangi, hessian
```

```python
img = io.imread("images/leaf.jpg")
img = rgb2gray(img)
```

```python
meijering_img = meijering(img)
sato_img = sato(img)
frangi_img = frangi(img)
hessian_img = hessian(img)
```

```python
fig = plt.figure(figsize=(20,20))
ax1 = fig.add_subplot(3,1,1)
ax1.imshow(img, cmap='gray')
ax1.title.set_text("input image")

ax1 = fig.add_subplot(3,2,3)
ax1.imshow(meijering_img, cmap='gray')
ax1.title.set_text("meijering image")

ax1 = fig.add_subplot(3,2,4)
ax1.imshow(sato_img, cmap='gray')
ax1.title.set_text("sato image")

ax1 = fig.add_subplot(3,2,5)
ax1.imshow(frangi_img, cmap='gray')
ax1.title.set_text("frangi image")

ax1 = fig.add_subplot(3,2,6)
ax1.imshow(hessian_img, cmap='gray')
ax1.title.set_text("hessian image")

plt.show()
```
