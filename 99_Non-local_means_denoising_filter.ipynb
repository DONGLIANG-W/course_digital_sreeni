## Non-local means (NLM) denoising filter

As the name of the algothm says, means indicates it calculate the mean of a set of pixel.

Non-local indicates it looks not only the local, the family of weights depend on the similarity between the pixels i and j. in the other words, similar pixel neighborhoods give larger weights.




```python
import cv2
import numpy as np
from skimage import io, img_as_float
from skimage.restoration import denoise_nl_means, estimate_sigma
```

```python
img = img_as_float(io.imread("images/BSE_Google_noisy.jpg"))
```

```python
sigma_est = np.mean(estimate_sigma(img, channel_axis=False))
```

```python
denoise_img = denoise_nl_means(img, h=1.15*sigma_est, fast_mode=True, 
                              patch_size=5, patch_distance=3)
```

```python
io.imshow(denoise_img)
```
