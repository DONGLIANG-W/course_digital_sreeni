```python
from skimage import io, filters, feature
from skimage.filters import roberts, sobel, scharr, prewitt, farid
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
img = cv2.imread("images/BSE_Google.jpg",0)
```

```python
io.imshow(img)
```

```python
roberts_img = roberts(img)
sobel_img = sobel(img)
scharr_img = scharr(img)
prewitt_img = prewitt(img)
farid_img = farid(img)
```

```python
io.imshow(roberts_img)
```

```python
io.imshow(sobel_img)
```

```python
io.imshow(scharr_img)
```

```python
io.imshow(prewitt_img)
```

```python
io.imshow(farid_img)
```

```python
canny_edge = cv2.Canny(img, 50, 80)
```

```python
io.imshow(canny_edge)
```

```python
# auto canny
sigma = 0.3
median = np.median(img)
lower = int(max(0,(1.0-sigma)*median))
upper = int(min(255,(1.0+sigma)*median))
auto_canny = cv2.Canny(img, lower, upper)
io.imshow(auto_canny)
```
