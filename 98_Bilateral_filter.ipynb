## Bilateral filter is a edge preserving denoising filter

Replaces each pixel value with a weighted average of nearby pixel values but takes into account the variation of intensities to preserve edges


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage import io, img_as_float
from skimage.restoration import denoise_bilateral
```

```python
img_gaussian_noise = cv2.imread("images/BSE_Google_noisy.jpg",0)
img = img_gaussian_noise
```

```python
bilateral_cv2 = cv2.bilateralFilter(img, 5, 20, 100, borderType=cv2.BORDER_CONSTANT)
```

```python
bilateral_skimage = denoise_bilateral(img, sigma_color=0.05, sigma_spatial=15)
```

```python
io.imshow(img)
```

```python
io.imshow(bilateral_cv2)
```

```python
io.imshow(bilateral_skimage)
```
