## what unsharp mask is and how to unsharp an image


```python
from skimage import io, img_as_float
from skimage.filters import unsharp_mask
from skimage.filters import gaussian
from matplotlib import pyplot as plt
```

```python
img = img_as_float(io.imread("images/BSE_Google_noisy.jpg"))
```

```python
gaussian_img = gaussian(img, sigma=1, mode='constant', cval=0.0)
```

```python
img2 = (img - gaussian_img)*1
```

```python
img3 = img + img2
```

```python
plt.imshow(img2, cmap='gray')
```

```python
plt.imshow(img3, cmap='gray')
```

```python
img = io.imread("images/BSE_Google_noisy.jpg")
```

```python
unsharp_img = unsharp_mask(img, radius=3, amount=1.0)
```

```python
fig = plt.figure(figsize=(12,12))
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(img, cmap='gray')
ax1.title.set_text("input image")
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(unsharp_img, cmap='gray')
ax2.title.set_text('unsharp image')
plt.show()
```
