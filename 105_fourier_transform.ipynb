## Fourier Transform in image processing

```python
from skimage import io
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

#### Generate a 2D sine wave image

```python
x = np.arange(256)
y = np.sin(2*np.pi*x/60)
y += max(y)

img = np.array([[y[j]*127 for j in range (256)] for i in range (256)], dtype=np.uint8)
plt.imshow(img)
```

```python
plt.plot(x,y)
```

Create a variable dft to store the Fourier transform data, note that the generate matrix is two channels, one for real and second for imaginery part

```python
dft = cv2.dft(np.float32(img), flags=cv2.DFT_COMPLEX_OUTPUT)
```

calculate the magnitude of the spectrum

```python
magnitude_spectrum = 20*np.log((cv2.magnitude(dft[:,:,0], dft[:,:,1]))+1)
```

```python
plt.imshow(magnitude_spectrum)
```

```python
dft_shift = np.fft.fftshift(dft)
```

```python
magnitude_spectrum = 20*np.log((cv2.magnitude(dft_shift[:,:,0], dft_shift[:,:,1]))+1)
```

```python
plt.imshow(magnitude_spectrum)
```
