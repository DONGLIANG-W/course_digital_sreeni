### Continue from the the last section, filter image in frequency domain

```python
from skimage import io
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
img = io.imread('images/BSE_Google.jpg')
```

```python
img.shape
```

```python
dft = cv2.dft(np.float32(img), flags=cv2.DFT_COMPLEX_OUTPUT)
```

```python
dft.shape
```

```python
dft_shift = np.fft.fftshift(dft)
```

get the magnitude of the spectrum

```python
magnitude_spectrum = 20*np.log(cv2.magnitude(dft_shift[:,:,0], dft_shift[:,:,1]))
```

```python
magnitude_spectrum
```

Create a high pass filter mask

```python
rows, cols = img.shape
crow, ccol = int(rows / 2), int(cols / 2)

mask = np.ones((rows, cols, 2), np.uint8)
r = 15 # create a circular shape mask, with radius r
center = [crow, ccol]
x, y = np.ogrid[:rows, :cols]
mask_area = (x - center[0]) ** 2 + (y - center[1]) ** 2 <= r*r
mask[mask_area] = 0
```

```python
fshift = dft_shift * mask
```

After multipling with the mask, do the reverse shift, and inverse Fourier transform, to get the image in spatial domain

```python
fshift_mask_mag = 2000*np.log(cv2.magnitude(fshift[:,:,0], fshift[:,:,1])+1) #this is only for viewing
```

```python
fshift_mask_mag.min()
```

```python
io.imshow(fshift_mask_mag)
```

```python
f_ishift = np.fft.ifftshift(fshift)
img_back = cv2.idft(f_ishift)
img_back = cv2.magnitude(img_back[:,:,0], img_back[:,:,1])
```

```python
fig = plt.figure(figsize = (20,20))
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(img, cmap='gray')
ax1.title.set_text('input image')

ax1 = fig.add_subplot(2,2,2)
ax1.imshow(magnitude_spectrum)
ax1.title.set_text('FFT of image')

ax1 = fig.add_subplot(2,2,3)
ax1.imshow(fshift_mask_mag)
ax1.title.set_text('FFT+MASK image')

ax1 = fig.add_subplot(2,2,4)
ax1.imshow(img_back,cmap='gray')
ax1.title.set_text('after inverse FFT')
```
