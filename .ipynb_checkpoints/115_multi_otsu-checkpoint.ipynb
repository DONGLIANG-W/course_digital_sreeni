### Otsu generate binary image with one threshold. Multi-Otsu enable multiple threshold value


```python
import matplotlib.pyplot as plt
import numpy as np
from skimage import io, img_as_ubyte, data
from skimage.filters import threshold_multiotsu
```

```python
img = io.imread("images/BSE_Google.jpg")
```

```python
io.imshow(img)
```

```python
thresholds = threshold_multiotsu(img, classes = 5)
```

```python
thresholds
```

### as shown above, the second input is the number of classes that user like to have to separate the image, here we give 5, and the output generated four threshold values

```python
plt.hist(img, bins=100)
```

```python
regions = np.digitize(img, bins=thresholds)
```

#### The np.digitize method use the threshold values to get generate the a categorized image, if there is only one threshold, it is a binary image, but here are four threshold, the image has 5 levels of brightness

```python
output = img_as_ubyte(regions)
```

```python
io.imshow(output)
```

```python
plt.hist(output)
```
